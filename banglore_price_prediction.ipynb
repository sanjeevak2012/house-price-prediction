```python
# we have an numerical column with missing values , how do you handle it, KNNimputer is one of the ways
# we have non-numerical column with missing values, how do you handle it, SimpleImputer
from sklearn.impute import KNNImputer, SimpleImputer
df_imputed = df2.copy()
knn_imputer_bath = KNNImputer(n_neighbors=5) # 5 

# use knn imputer and put the imputed/predicted/average/mean/median values in tha bath column 
# where ever the values are missing
df_imputed['bath'] = knn_imputer_bath.fit_transform(df_imputed[['bath']])
```

```python
# For categorical columns, 'most_frequent' (mode) is a common strategy.
simple_imputer_size = SimpleImputer(strategy='most_frequent')
df_imputed['size'] = simple_imputer_size.fit_transform(df_imputed[['size']])
```
